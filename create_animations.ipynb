```python
import generate_data
import objects_utils
import gl_graph

import graphlearning
from scipy import sparse
import numpy as np
import random
```

# MNIST Uniform kernel

```python
# Load in VAE MNist data from graphlearning library
X,y = graphlearning.datasets.load("mnist", "vae")
label_arr = [4, 9]
n = 20

X, y, W = generate_data.GenerateWMat(X, y, n, label_arr)

# Save data
sparse.save_npz("data/X_mnist.npz", sparse.csr_matrix(X))
sparse.save_npz("data/W_mnist_uniform.npz", W)
np.save('data/y_mnist.npy', list(y))

evals_arr, evecs_arr = generate_data.GenEvecs(W)
np.save('data/ray_mnist_uniform.npy', evals_arr)
np.save('data/evecs_mnist_uniform.npy', evecs_arr)
```

```python
evecs_arr = np.load('data/evecs_mnist_uniform.npy')
evals_arr = np.load('data/ray_mnist_uniform.npy')
y         = np.load('data/y_mnist.npy')

objects_utils.save_animation(
    evec_seq=evecs_arr,
    eigenvalue_seq=evals_arr,
    labels=y,
    title="MNIST Uniform",
    path="animation_mnist_uniform"   # writes animation.html + animation_data.js
)
```
